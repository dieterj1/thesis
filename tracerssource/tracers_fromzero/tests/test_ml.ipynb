Reading in the test matched file

```python
infile = r"/home/toon/Documents/PHD/coding/pyrust-docker/tracers/tests/testout/_0.csv"

#even lines are the matched points
#odd lines are the original input

matches = []
ogpoints = []

with open(infile, "r") as f:
    for matchline in f:
        matchline = matchline.rstrip()  # remove newline character
        if matchline:
            # process matchline, it has format int,float,float,int|...
            strtupls = matchline.split("|")
            matches.append([(int(split[0]),float(split[1]),float(split[2]),int(split[3]))for split in [x.split(",") for x in strtupls]])

        # origline
        origline = f.readline()
        #strip
        origline = origline.rstrip()
        if origline:
            # process origline, it has format float,float,int|...
            split = origline.split(",")
            ogpoints.append((float(split[0]),float(split[1]),int(split[2])))
```

```python
match_nodeids = [[mypygraph.osm_to_nodeid(
    a) for a, b, c, d in match] for match in matches]
print(match_nodeids)
match_nodeids = [[a for a in b if a != None] for b in match_nodeids]
match_nodeids

og_lonlats = [(a, b) for a, b, _c in ogpoints]
```

```python
res = mypygraph.extract_features_labeled_pert(match_nodeids, og_lonlats)
# convert to pd df
import pandas as pd
df = pd.DataFrame(res, columns=["step_d","step_bear","roadtype","emit_a_d","emit_a_bear","emit_b_d","emit_b_bear","y"])
df
```
